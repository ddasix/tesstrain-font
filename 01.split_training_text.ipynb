````python
import os
import random
import pathlib
import subprocess
````

````python
training_text_file = 'langdata_lstm/kor/kor.training_text'
````

````python
lines = []
````

````python
with open(training_text_file, 'r') as input_file:
    for line in input_file.readlines():
        lines.append(line.strip())
````

````python
output_directory = 'tesstrain/data/NanumGothic-ground-truth'
````

````python
if not os.path.exists(output_directory):
    os.makedirs(output_directory)
````

````python
random.shuffle(lines)
````

````python
count = 10
````

````python
lines = lines[:count]
````

````python
line_count = 0
````

````python
for line in lines:
    training_text_file_name = pathlib.Path(training_text_file).stem
    line_training_text = os.path.join(output_directory, f'{training_text_file_name}_{line_count}.gt.txt')

    with open(line_training_text, 'w') as output_file:
        output_file.writelines([line])

    file_base_name = f'kor_{line_count}'

    subprocess.run([
        'text2image',
        '--font=NanumGothic',
        f'--text={line_training_text}',
        f'--outputbase={output_directory}/{file_base_name}',
        '--max_pages=1',
        '--strip_unrenderable_words',
        '--leading=32',
        '--xsize=3600',
        '--ysize=480',
        '--char_spacing=1.0',
        '--exposure=0',
        '--unicharset_file=langdata_lstm/kor/kor.unicharset'
    ])

    line_count += 1
````
